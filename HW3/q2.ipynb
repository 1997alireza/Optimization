#Log barrier method

Log barrier method applied on a convex quadratic function subject to a random linear inequality using different `mu`

```python
from HW3.log_barrier import barrier_method
from HW3.matrices import *
from HW3.function_provider import QuadraticFunctionProvider
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
n = 100
b = np.ones(n)
P = np.random.rand(n, n)
q = np.random.rand(n, 1) + np.ones([n, 1])
```

```python
A = tridiag(size=n)
    
quadratic_provider = QuadraticFunctionProvider(A, b)

for mu in [1.5, 3, 5, 10]:
    
    _, newton_iters, gaps = barrier_method(quadratic_provider, P, q, mu=mu)
    plt.plot(newton_iters, gaps, label=mu)

plt.legend(loc='lower left')
plt.yscale('log')
plt.xlabel('Newton Iterations')
plt.ylabel('Duality gap')
plt.title('Log barrier method applied on a convex quadratic function subject to a random linear inequality '
          '(on different `mu`)')
plt.show()
```

```python
A = hilb(size=n)
    
quadratic_provider = QuadraticFunctionProvider(A, b)

for mu in [1.5, 3, 5, 10]:
    
    _, newton_iters, gaps = barrier_method(quadratic_provider, P, q, mu=mu)
    plt.plot(newton_iters, gaps, label=mu)

plt.legend(loc='lower left')
plt.yscale('log')
plt.xlabel('Newton Iterations')
plt.ylabel('Duality gap')
plt.title('Log barrier method applied on a convex quadratic function subject to a random linear inequality '
          '(on different `mu`)')
plt.show()
```
